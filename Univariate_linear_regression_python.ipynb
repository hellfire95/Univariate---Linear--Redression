## ML MODEL
___


### Hypothesis

<center>$h_\theta(x) = \theta.x$

### Cost Function
<center>C= $\frac{1}{2m}$ $\sum_{i=1}^{m} (h_\theta(x)^i-y^i)^2$

```python
def cost(x,y, theta):
    hypo=np.dot(x,theta)
    square=(hypo-y)**2
    cost_value=(1/(2*m))*(np.sum(square))
    return cost_value
```

```python
cost(x,y,theta)
```

### Gradient Descent
___

<center>$\theta_j=\theta_j - \frac{1}{m}\alpha\sum_{i=1}^{m}(h_\theta(x)^i-y^i)x^i $

```python
def grad_des(x,y, theta, alpha, epoch):
    costs=[]
    for i in range(epoch):
        hypo=np.dot(x,theta)
        minus=hypo-y
        all_but_theta=(alpha/m)*(np.sum(np.dot((np.transpose(x)),minus)))
        theta -= all_but_theta
        costs.append(cost(x,y,theta))
    return costs, theta
```

### Storing Cost and theta values

```python
costs, theta= grad_des(x,y, theta , 0.0003, 1000)
```

```python
costs
```

### Plotting Costs
